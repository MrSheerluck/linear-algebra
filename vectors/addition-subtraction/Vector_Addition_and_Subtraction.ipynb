```python
%%manim -ql VectorAdditionCleanAxes

from manim import *

class VectorAdditionCleanAxes(Scene):
    def construct(self):
        # Axes (no grid)
        axes = Axes(
            x_range=[0, 5, 1],
            y_range=[0, 5, 1],
            x_length=6,
            y_length=6,
            axis_config={"include_tip": True}
        )

        axes_labels = axes.get_axis_labels("x", "y")

        self.play(Create(axes), Write(axes_labels))

        # Vector values
        a = np.array([3, 1, 0])
        b = np.array([1, 3, 0])
        c = a + b

        # Vectors
        vec_a = Arrow(
            start=axes.c2p(0, 0),
            end=axes.c2p(3, 1),
            color=BLUE,
            buff=0
        )

        vec_b = Arrow(
            start=axes.c2p(3, 1),
            end=axes.c2p(4, 4),
            color=GREEN,
            buff=0
        )

        vec_c = Arrow(
            start=axes.c2p(0, 0),
            end=axes.c2p(4, 4),
            color=YELLOW,
            buff=0
        )

        # Coordinate labels
        coord_a = MathTex("(3,1)").next_to(vec_a.get_end(), RIGHT)
        coord_b = MathTex("(4,4)").next_to(vec_b.get_end(), RIGHT)
        coord_c = MathTex("(4,4)").next_to(vec_c.get_end(), UP)

        label_a = MathTex(r"\vec{a}=(3,1)").next_to(vec_a, UP)
        label_b = MathTex(r"\vec{b}=(1,3)").next_to(vec_b, RIGHT)
        label_c = MathTex(r"\vec{a}+\vec{b}=(4,4)").next_to(vec_c, UP)

        # Animate
        self.play(GrowArrow(vec_a), Write(label_a), Write(coord_a))
        self.wait(0.5)

        self.play(GrowArrow(vec_b), Write(label_b), Write(coord_b))
        self.wait(0.5)

        self.play(GrowArrow(vec_c), Write(label_c))
        self.wait(2)
```

```python
%%manim -ql VectorSubtractionNegation

from manim import *

class VectorSubtractionNegation(Scene):
    def construct(self):
        # Axes (clean, no grid)
        axes = Axes(
            x_range=[-1, 5, 1],
            y_range=[-3, 4, 1],
            x_length=6,
            y_length=6,
            axis_config={"include_tip": True}
        )

        axes_labels = axes.get_axis_labels("x", "y")
        self.play(Create(axes), Write(axes_labels))

        # Vector values
        a = np.array([3, 1, 0])
        b = np.array([1, 3, 0])
        neg_b = -b
        result = a + neg_b

        # Vectors
        vec_a = Arrow(
            axes.c2p(0, 0),
            axes.c2p(3, 1),
            color=BLUE,
            buff=0
        )

        vec_b = Arrow(
            axes.c2p(0, 0),
            axes.c2p(1, 3),
            color=GREEN,
            buff=0
        )

        vec_neg_b = Arrow(
            axes.c2p(3, 1),
            axes.c2p(2, -2),
            color=RED,
            buff=0
        )

        vec_result = Arrow(
            axes.c2p(0, 0),
            axes.c2p(2, -2),
            color=YELLOW,
            buff=0
        )

        # Labels
        label_a = MathTex(r"\vec a=(3,1)").next_to(vec_a, UP)
        label_b = MathTex(r"\vec b=(1,3)").next_to(vec_b, LEFT)
        label_neg_b = MathTex(r"-\vec b=(-1,-3)").next_to(vec_neg_b, RIGHT)
        label_result = MathTex(r"\vec a-\vec b=(2,-2)").next_to(vec_result, DOWN)

        coord_a = MathTex("(3,1)").next_to(vec_a.get_end(), RIGHT)
        coord_neg_b = MathTex("(2,-2)").next_to(vec_neg_b.get_end(), RIGHT)

        # Animation sequence
        self.play(GrowArrow(vec_a), Write(label_a), Write(coord_a))
        self.wait(0.5)

        self.play(GrowArrow(vec_b), Write(label_b))
        self.wait(0.5)

        # Transform b into -b
        self.play(Transform(vec_b, vec_neg_b), Write(label_neg_b))
        self.wait(0.5)

        # Resultant
        self.play(GrowArrow(vec_result), Write(label_result), Write(coord_neg_b))
        self.wait(2)
```
