```python
%%manim -qm -v WARNING VectorMultipleScaleUp
from manim import *

class VectorMultipleScaleUp(Scene):
    def construct(self):
        plane = NumberPlane(
            x_range=[-6, 6],
            y_range=[-4, 4],
        )
        self.add(plane)

        # Direction line (same slope for all)
        direction_line = Line(
            plane.c2p(-6, -3),
            plane.c2p(6, 3),
            color=YELLOW
        )
        self.play(Create(direction_line))

        # Base vector
        base_vec = Vector([2, 1], color=BLUE)
        label = MathTex(r"\vec{v}=(2,1)").next_to(base_vec, UP)

        self.play(Create(base_vec), Write(label))
        self.wait(1)

        # Scaling factors
        scales = [2, 3, 4]
        colors = [GREEN, ORANGE, RED]

        for k, color in zip(scales, colors):
            scaled_vec = Vector([2*k, 1*k], color=color)
            scaled_label = MathTex(
                rf"{k}\vec{{v}} = ({2*k},{1*k})"
            ).next_to(scaled_vec, UP)

            self.play(
                Transform(base_vec, scaled_vec),
                Transform(label, scaled_label),
                run_time=1.2
            )
            self.wait(0.7)

        self.wait(2)

```

```python
%%manim -qm -v WARNING VectorMultipleShrink
from manim import *

class VectorMultipleShrink(Scene):
    def construct(self):
        plane = NumberPlane(
            x_range=[-6, 6],
            y_range=[-4, 4],
        )
        self.add(plane)

        # Direction line
        direction_line = Line(
            plane.c2p(-6, -3),
            plane.c2p(6, 3),
            color=YELLOW
        )
        self.play(Create(direction_line))

        # Base vector
        base_vec = Vector([4, 2], color=BLUE)
        label = MathTex(r"\vec{v}=(4,2)").next_to(base_vec, UP)

        self.play(Create(base_vec), Write(label))
        self.wait(1)

        # Shrink factors
        scales = [1/2, 1/4, 1/8]
        colors = [GREEN, ORANGE, RED]

        for k, color in zip(scales, colors):
            scaled_vec = Vector([4*k, 2*k], color=color)
            scaled_label = MathTex(
                rf"\tfrac{{1}}{{{int(1/k)}}}\vec{{v}} = ({4*k},{2*k})"
            ).next_to(scaled_vec, UP)

            self.play(
                Transform(base_vec, scaled_vec),
                Transform(label, scaled_label),
                run_time=1.2
            )
            self.wait(0.7)

        self.wait(2)

```
